```python
from pyproj import Transformer
import pyproj
import numpy as np
import pandas  as pd
from shapely.geometry import Point as point
import geopandas as gpd
import folium
```

### 관악08

```python
bus = pd.read_csv('관악08 좌표.csv', encoding='cp949') # 해당 엑셀 파일 불러오기
bus
```

```python
bus_id = bus['표준버스정류장ID']
bus_x = bus['위도']
bus_y = bus['경도']
```

```python
# Case 2
'''
- x, y 좌표를 받아서 좌표계 변환 없이 m.html에 시각화하기 
'''
#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
lat = bus['위도'].mean()
long = bus['경도'].mean()

#지도 띄우기
m = folium.Map([lat, long], zoom_start = 9)

coords = []
for i in range(len(bus)-1):
    x = bus_x[i]
    y = bus_y[i]
    coords.append([x, y])
    
for i in range(len(coords)):
    folium.Circle(
        location = coords[i],
        radius = 50,
        color = '#191970',
        fill = 'crimson',
    ).add_to(m)

m.save('관악08.html')
```
